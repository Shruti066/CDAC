```python
df1=df.drop('Unnamed: 0', axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df1.model=le.fit_transform(df1.model)
```

```python
df1.transmission=le.fit_transform(df1.transmission)
```

```python
df1.fuelType=le.fit_transform(df1.fuelType)
```

```python
df1.Make=le.fit_transform(df1.Make)
```

```python
df1.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df1.drop('price', axis=1), df1.price, test_size=0.2, random_state=50)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_test=scaler.fit_transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr_pred=lr.predict(x_test)
lr_pred
```

```python
from sklearn import metrics
from sklearn.metrics import r2_score, mean_squared_error
```

```python
metrics.mean_squared_error(y_test, lr_pred)
```

```python
metrics.r2_score(y_test, lr_pred)
```

```python
import sklearn
from sklearn import tree
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor()
```

```python
rfr.fit(x_train,y_train)
```

```python
y_pred=rfr.predict(x_test)
```
